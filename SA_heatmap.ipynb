```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

csv_file = '/content/drive/MyDrive/Colab Notebooks/박논/pandas/sachain0504.csv'

# csv 파일을 읽고, 데이터프레임으로 변환
df = pd.read_csv(csv_file)
df.head()
```

```python
# Speech_act1과 Speech_act2를 기준으로 데이터 집계
pivot_table = pd.pivot_table(df, index='SpeechAct1', columns='SpeechAct2', aggfunc=len, fill_value=0)

# 히트맵 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap="Blues", fmt='d', cbar=False)
plt.title("Speech Act Heatmap")

# 히트맵을 png 이미지로 저장
# bbox_inches='tight'를 통해 이미지의 여백 최소화
plt.savefig("sa0504_heatmap.png", dpi=300, bbox_inches='tight')

plt.show()
```
