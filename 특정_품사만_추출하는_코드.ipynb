# 개념집합 출력

입력 문장 Sentence에서 형태소 태그에서 명사(Noun) 동사(Verb(, 형용사(Adjective)만 추출하는 코드

```python
from konlpy.tag import Okt

okt = Okt()

# 입력 문장
sentence = "밖/Noun,에/Josa,나오다/Verb,./Punctuation"

# 형태소 분석 후 품사가 'Noun', 'Verb', 'Adjective'인 형태소 추출
concepts = []
for word in sentence.split(','):
    stem, postag = word.split('/')
    if postag in ['Noun', 'Verb', 'Adjective']:
        concepts.append(f"{stem}/{postag}")

# 결과 출력
print(", ".join(concepts))
```

# 데이터를 불러 들여서 개념 집합을 추출한 후 결과를 출력

- word 변수를 슬래시를 포함하는 단어와 그렇지 않은 단어로 구분하기 위해 word_parts 변수에 word.split('/') 결과를 할당.  
- len(word_parts)를 사용하여 word_parts에 2개의 값이 있는 경우에만 stem과 postag 변수에 할당 
- 값이 없는 경우는 추출된 형태소가 없습니다라고 출력
- 파일을 prin()하여 보고 주소 쓰기로 저장


```python
from konlpy.tag import Okt

okt = Okt()

# 입력 파일에서 문장 읽어오기
with open('/content/drive/MyDrive/Colab Notebooks/mess_pos/messengser_okt.csv', 'r', encoding='utf-8') as f:
    lines = f.readlines()

# 결과를 저장할 파일 열기
with open('output.txt', 'w', encoding='utf-8') as f:
    # 각 라인에 대해 형태소 분석 후 품사가 Noun, Verb, Adjective인 형태소 추출
    for line in lines:
        sentence = line.strip()  # 개행문자 제거
        concepts = []
        for word in sentence.split(','):
            word_parts = word.split('/')
            if len(word_parts) == 2:  # 슬래시를 포함하는 경우에만 처리
                stem, postag = word_parts
                if postag in ['Noun', 'Verb', 'Adjective']:
                    concepts.append(f"{stem}/{postag}")
        
        # 결과 출력
        if concepts:
            result = ", ".join(concepts)
            print(result)
            f.write(result + '\n')  # 파일에도 쓰기
        else:
            print("추출된 형태소가 없습니다.")
            f.write("추출된 형태소가 없습니다.\n")  # 파일에도 쓰기
```

- 위 코드를 실행하면 입력 파일의 각 라인에 대해 형태소 분석을 수행, 
- 품사가 동사, 형용사, 명사인 형태소만 추출하여 print() 함수로 출력 
- output.txt 파일에도 결과 저장
